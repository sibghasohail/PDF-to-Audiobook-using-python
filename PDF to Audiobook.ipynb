```python

import fitz  # PyMuPDF
from gtts import gTTS

def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF file."""
    text = ""
    pdf_document = fitz.open(pdf_path)
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    return text

def text_to_speech(text, audio_path):
    """Convert text to speech and save as an audio file."""
    tts = gTTS(text=text, lang='en')  # Change 'en' to your preferred language if needed
    tts.save(audio_path)

def pdf_to_audio(pdf_path, audio_path):
    """Convert a PDF file to an audio file."""
    pdf_text = extract_text_from_pdf(pdf_path)
    text_to_speech(pdf_text, audio_path)

# Example usage
pdf_path = 'Sample Form for Students.pdf'  # Replace with the path to your PDF
audio_path = 'output.mp3'       # Desired output audio file path

pdf_to_audio(pdf_path, audio_path)
```
